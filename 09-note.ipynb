```python
import random
while True:
  rand_num = random.randint(1,6)
  print(rand_num)
  if rand_num == 4:
    break
  print('.')
```

trueは無限ループ
無限ループをとめるにはbreak
4が出たら無限ループを止めるシステム
最後のピリオドなしはbreakがあるとそのあとのやつを無視する。１０６

```python
import random
while True:
  rand_num = random.randint(1,6)
  print(rand_num)
  if rand_num == 4:
    continue
  print('.')
```
